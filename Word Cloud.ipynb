Word Cloud

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

Cargamos los datasets 'cursos_coursera', 'edx' y 'udemy' en DataFrames

```python
cursos_coursera = pd.read_csv('data/Coursera_courses.csv')
price_level_coursera = pd.read_csv('data/coursera_price_level.csv')
coursera_reviews = pd.read_csv('data/Coursera_reviews.csv')
edx = pd.read_csv('data/edx_courses.csv')
udemy = pd.read_csv('data/udemy_courses.csv') 
```

combinamos los t√≠tulos de los cursos en una sola lista

```python
titles = cursos_coursera["name"].tolist() + edx["title"].tolist() + udemy["course_title"].tolist() + price_level_coursera["title"].tolist()
```

```python

# unimos los t√≠tulos de los cursos en una sola cadena de texto
text = " ".join(titles)

# creamos la nube de palabras
wordcloud = WordCloud().generate(text)

# mostramos la nube de palabras
plt.imshow(wordcloud, interpolation='mitchell')
plt.axis("off")
plt.show()
```
